```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow import keras
```

```python
df = pd.read_csv('diabetes.csv')
df
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
                                                      'BMI', 'DiabetesPedigreeFunction', 'Age']], df.Outcome, test_size=0.2, 
random_state=25)
```

```python
len(X_train)
```

```python
X_train_scaled = X_train.copy()
X_train_scaled['Age'] = X_train_scaled['Age']/100

X_test_scaled = X_test.copy()
X_test_scaled['Age'] = X_test_scaled['Age']/100
```

```python
model = keras.Sequential([
    keras.layers.Dense(1, input_shape=(8,), activation='sigmoid', kernel_initializer='ones', bias_initializer='zeros')
   
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(X_train_scaled, y_train, epochs=600)
```

```python
model.evaluate(X_test_scaled, y_test)
```
